### The Self-Organising Map (SOM): 
* unsupervised machine learning algorithm
* the map organises itself without any instruction from others
* trained through a competitive neural network, a single-layer feed-forward network.

##### SOM can also be considered a type of manifold learning that 
* handles non-linear dimensionality reduction
* represent high-dimensional observable data onto lower dimension latent space, typically on a 2D square grid
* preserves the topology of the original input space.

#### The neural network of SOM has:
* one input layer
* one output layer
* The second layer is a 2D lattice of m x n lattice points
* Each lattice point is densely connected to all points in the input layer, possessing different weight values.

#### 4 steps for SOM:
1. Initialisation:
   Weights of lattice points in the map layer are initialised.
2. Competitive process:
   Select one input sample and search the best matching unit among all lattice points in n x m grid using distance measures.
3. Cooperative process:
   Find the proximity lattice points of BMU by neighbourhood function.
4. Adaptation process:
   Update the BMU and neighbours' weights by shifting the values towards the input pattern.

#### The competitive process:
* to search for the most similar lattice point with the input pattern; the winner is called Best Matching Unit (BMU). We calculate the Euclidean distance and the grid point with the shortest distance from the input feature becomes the BMU
* the BMU learns together with the neighbouring lattice points (neighbourhood is defined by the designer)
* small neighbourhood leads to overfitting, risk of having some dead lattice points (never have an opportunity to learn)

#### Advantages of SOM:
* useful for reducing the dimensionality of data and visualizing high-dimensional data in a low-dimensional (typically 2D) space. This can help in understanding the structure and distribution of the data.
* SOMs are an unsupervised learning technique, meaning they do not require labeled data. This is advantageous when labeled data is scarce.
* SOMs can automatically detect patterns and clusters in the data, which can be useful for exploratory data analysis.
* One of the key features of SOMs is that they preserve the topological properties of the input space. Similar input data points are mapped to nearby nodes in the SOM, which helps in maintaining the inherent structure of the data.
* SOMs can capture nonlinear relationships in the data, making them suitable for more complex datasets.


The learning rate and neighborhood size are critical hyperparameters in the training of Self-Organizing Maps (SOMs), and they significantly influence the performance and outcome of the algorithm. Here are the effects of these parameters:

##### 1. Learning Rate:
* The learning rate determines the extent to which the weights of the SOM nodes are updated during each training step.
* High learning rate: faster convergence leaving to instability; less accurate map as the map can overshoot optimal weight adjustments.
* Low learning rate:  Leads to slower convergence, providing a more stable training process but requiring more iterations to reach an optimal state; more accurate map as smaller adjustments allow the weights to fine-tune more precisely to the data.

##### 2. Neighbourhood Size:
* The neighborhood size determines the radius within which the neighboring nodes around the Best Matching Unit (BMU) are updated during training.
* Large Size: promotes global topological preservation early in training; achieves faster convergence to a rough map structure.
* Small size: promotes local topological preservation and important for fine tuning the map; leads to more distinct and precise clusters, as fewer nodes are updated simultaneously, allowing for finer differentiation; better representation of data's finer details.
